2. Write a tokenization function to process our text data

Before we can classify any posts, we need to clean and `tokenize` the text data. Here, we will implement a tokenize function. This function will perform the following steps on the string, `text`, using nltk:

1. Identify any urls in text, and replace each one with the word, "urlplaceholder".
2. Split text into tokens.
3. For each token: lemmatize, normalize case, and strip leading and trailing white space.
4. Return the tokens in a list!

```python
url_regex = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'

def tokenize(text):
    # get list of all urls using regex
    detected_urls = re.findall(url_regex, text)
     # replace each url in text string with placeholder
    for url in detected_urls:
        text = text.replace(url, 'urlplaceholder')
    
    # normalize, tokenize, remove punctuation and remove leading/trailing white spaces
    tokens = nltk.word_tokenize(re.sub(r"[^a-zA-Z0-9]", " ", text.lower().strip()))
    
    # remove stopwords
    tokens = [txt for txt  in tokens if txt not in stopwords.words('english')]

    # lemmatize
    lemmatizer=WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    
    return tokens
```

## 3. Build a machine learning pipeline

This pipeline should take in the message column as input and output classification results on the other 36 categories in the dataset. We can use MultiOutputClassifier which is helpful for predicting multiple target variables.

Let's first train Decision tree classifier 

Let's use bag_of_words and TFIDF to extract features from our text and using Decision Tree Classifier

```python
X, y = load_data()

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size= 0.25, random_state= 42)

# Instantiate tranformers and classifiers
vect = CountVectorizer(tokenizer=tokenize)
tfidf = TfidfTransformer()
clf = DecisionTreeClassifier(max_depth=5)

# fit and tranform the training data
X_train_counts = vect.fit_transform(X_train)
X_train_tfidf = tfidf.fit_transform(X_train_counts)

#train the classifier
clf.fit(X_train_tfidf, y_train)

# predict on test data
X_test_counts = vect.transform(X_test)
X_test_tfidf = tfidf.transform(X_test_counts)
y_pred = clf.predict(X_test_tfidf)
```

Let's look at the results

```python
labels = np.unique(y_pred)
confusion_mat = confusion_matrix(y_test.values.argmax(axis=1), y_pred.argmax(axis=1), labels=labels)
accuracy = (y_pred == y_test).mean()

print("Labels:", labels)
print("Confusion Matrix:\n", confusion_mat)
print("Accuracy:", accuracy)

#Calculate the accuracy for each of them.
for i in range(len(y.columns)):
  print('Category: {} '.format(y.columns[i]))
  print(classification_report(y_test.iloc[:, i].values, y_pred[:, i]))
  print('Accuracy {}\n\n'.format(accuracy_score(y_test.iloc[:, i].values, y_pred[:, i])))
  print('F1 {}\n\n'.format(f1_score(y_test.iloc[:, i].values, y_pred[:, i],average='weighted')))

print(classification_report(y_test.iloc[:, 0:].values, np.array([x[0:] for x in y_pred]), target_names = y.columns))
```

###Let us rewrite the above code for TFIDF to use sklearn's `Pipeline`. 

Fortunately we can automate all the fitting, transforming, and predicting by chaining these estimators together into a single estimator object

```python
# load data from database
engine = create_engine('sqlite:///disaster.db')
df = pd.read_sql_table(table_name='my_disaster_response_table', con=engine)
X = df['message']
Y = df.iloc[:,4:]
```

```python
url_regex = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'

def tokenize(text):
    # get list of all urls using regex
    detected_urls = re.findall(url_regex, text)
     # replace each url in text string with placeholder
    for url in detected_urls:
        text = text.replace(url, 'urlplaceholder')
    
    # normalize, tokenize, remove punctuation and remove leading/trailing white space
    tokens = nltk.word_tokenize(re.sub(r"[^a-zA-Z0-9]", " ", text.lower()))
    
    # remove stopwords
    tokens = [txt for txt  in tokens if txt not in stopwords.words('english')]

    # lemmatize
    lemmatizer=WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    
    return tokens
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,random_state = 42)
```

```python
pipeline = Pipeline([
    ('vect', CountVectorizer(tokenizer = tokenize))
    , ('tfidf', TfidfTransformer())
    , ('clf', MultiOutputClassifier(DecisionTreeClassifier()))])
```

```python
pipeline.get_params().keys()
```

```python
%%time
pipeline.fit(X_train, Y_train)
```

```python
Y_pred = pipeline.predict(X_test)
    
# Calculate the accuracy for each of them.
for i in range(len(Y.columns)):
    print('Category: {} '.format(Y.columns[i]))
    print(classification_report(Y_test.iloc[:, i].values, Y_pred[:, i]))
    print('Accuracy {}\n\n'.format(accuracy_score(Y_test.iloc[:, i].values, Y_pred[:, i])))
    print('F1 {}\n\n'.format(f1_score(Y_test.iloc[:, i].values, Y_pred[:, i],average='weighted')))
```

```python
labels = np.unique(Y_pred)
confusion_mat = confusion_matrix(y_test.values.argmax(axis=1), Y_pred.argmax(axis=1), labels=labels)
accuracy = (Y_pred == Y_test).mean()

print("Labels:", labels)
print("Confusion Matrix:\n", confusion_mat)
print("Accuracy:", accuracy)
```

```python
print(classification_report(Y_test.iloc[:, 0:].values, np.array([x[0:] for x in Y_pred]), target_names = Y.columns))
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
class StartingVerbExtractor(BaseEstimator, TransformerMixin):

    def starting_verb(self, text):
        sentence_list = nltk.sent_tokenize(text)
        for sentence in sentence_list:
            pos_tags = nltk.pos_tag(tokenize(sentence))
            first_word, first_tag = pos_tags[0]
            if first_tag in ['VB', 'VBP'] or first_word == 'RT':
                return True
        return False

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X_tagged = pd.Series(X).apply(self.starting_verb)
        return pd.DataFrame(X_tagged)
```
